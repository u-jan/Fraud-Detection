We don't have a target yet. In other words we don't have a label to refer if an event is fraud or not and we have nothing to predict within our recent dataset and that should certainly change.

What could be done? After studying feature values within the dataset, we can use acct_type to refer.

```python
df.acct_type.value_counts()
```

Someone labelled the events in a different format. Now we can create our target based on that.

Following code checks if the word 'fraud' is included in any row of the column acct_type. If such word is included it will assign 1 to target which stands for a fraud event. Else, the event is not fraud.

```python
df['target'] = ['fraud' in val for val in df.acct_type]
```

```python
df.target[0:3] # looks like it worked 1= True, 0 = False
```

```python
df["target"] =  df["target"].map({False:0,True:1}) # but I like it this way. Its a style issue.
```

```python
df.target.head() #also type is int now.
```
